```python
# Compute correlation matrix
correlation_matrix = data.corr(method='pearson')   # Pearson, Kendall, Spearman

# Display correlation matrix
correlation_matrix
```

```python
## FROM CHATGPT, March 05,2025, 6:01 PM
# Find the most correlated pairs (excluding self-correlation)
correlation_unstacked = correlation_matrix.unstack().sort_values(ascending=False)

# remove corellation between a column and itself
most_correlated_pairs = correlation_unstacked[correlation_unstacked.index.get_level_values(0) != correlation_unstacked.index.get_level_values(1)]

# Display top correlated pairs
#print("\nMost correlated column pairs:")
most_correlated_pairs
```
